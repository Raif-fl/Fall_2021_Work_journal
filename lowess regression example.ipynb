```python
import numpy as np 
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
x=[0]*100
y=[0]*100
for i in range(100):
    x[i]=1+np.random.random()-0.5
    
for i in range(100):
    y[i]=1+np.random.random()-0.5
```

```python
plt.scatter(x,y)
plt.show()
```

```python
smoothed = statsmodels.nonparametric.smoothers_lowess.lowess(x, y, frac=0.5, it=3, delta=0.0, xvals=None, is_sorted=False, missing='drop', return_sorted=True)
```

```python
plt.scatter(np.hsplit(smoothed,2)[0],np.hsplit(smoothed,2)[1])
plt.ylim(0,2)
plt.plot()
```
