```python
from numpy.random import default_rng
rng = default_rng()
```

```python
x = rng.standard_normal(50)+5*30
y = rng.standard_normal(450)+5
z = rng.standard_normal(500)+5*20
```

```python
position = list(range(500))
```

```python
xy = np.concatenate((x,y))
```

```python
smoothed_vec_m, cumul_m = kat.get_smoothed_vector(position, xy)
smoothed_vec_c, cumul_c = kat.get_smoothed_vector(position, z)
```

```python
#figure(figsize = (12,6))
plt.plot(cumul_m)
plt.plot(cumul_c)
plt.ylabel("Cumulative Sum")
plt.xlabel("Transcript Position")
plt.title("Early Pause Site")
```
