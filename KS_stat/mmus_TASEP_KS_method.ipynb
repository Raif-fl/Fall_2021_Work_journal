## Define functions

```python
# Calculate the smoothed density vector pbar for xth entry with length n-9
def get_pbar_from_count(transcript):
    transcript = transcript + 1 #All elongation rates were calculated this way to set a new baseline
    M = sum(transcript)
    p = transcript/M
    x=0
    pbar=[]
    for px in p:
        pbar_x = 0.1*sum(p[x:x+10]) #it is x+10 not x+9 because python does not include the final index.
        pbar.append(pbar_x)
        x = x+1
        if x  == len(p)-9:
            break
    pbar = np.array(pbar)
    return(pbar)
```

### Step 5
We are going to filter out any count arrays that are of an insufficient length or have insufficient read density just like we did when calculating the KS statistic. We are also going to filter out any transcripts that have negative elongation rates. Negative elongation rates can sometimes occur when the particle density of a transcript is predicted to be too high. These negative values will interfere with our calculations so they are removed. Save the Gene ID's for the filtered count arrays into a list.  

### Step 6
For each transcript that passes our filter we are going to calculate 3 particle densities. The first is an altered particle density that is calculated by inducing elongation limitation in the control using the alter_p() function. The second and third are the particle densities of the mutant and control samples which can be easily calculated using the get_pbar_from_count function.

```python
# Find the correlation between all of the controls and the samples.
# Note that many of the lowest values spring from very short transcripts or transcripts with few reads. 
mutant_list = []
control_list = []
alt_p_list = []
pbar_m_list = []
pbar_c_list = []
gene_list = []
for tr_m, tr_c, el_m, el_c, gene_name in zip(data_mutant, data_control, elon_mutant, elon_control, names_mutant):
    if len(tr_m) > 100 and sum(tr_m)/len(tr_m) > 0.5 and sum(tr_c)/len(tr_c) > 0.5 and sum(el_c) > 0 and sum(el_m) > 0:
        mutant_list.append(tr_m)
        control_list.append(tr_c)
        alt_p_list.append(kat.alter_p(el_c, el_m))
        pbar_m_list.append(get_pbar_from_count(tr_m))
        pbar_c_list.append(get_pbar_from_count(tr_c))
        gene_list.append(gene_name)
```

### Step 7
Smooth out the particle densities using Loess smoothing and calculate the cumulative read distributions. the get_smoothed_vector function from the keegan\_analysis\_tools python file is set up to perform Loess smoothing with a window size equal to 5% of the transcript length and calculates the cumulative read distribution. This step can take some time so it is recommended to use the multiprocess.Pool() function to complete this using multiple cores. 

```python
max_pool = 20

if pre_load == False:
    with Pool(max_pool) as p:
        pool_alt_p = list(
            tqdm(
                p.imap(kat.get_smoothed_vector,
                       alt_p_list),
                total=len(alt_p_list)
            )
        )    
```

```python
if pre_load == False:
    with Pool(max_pool) as p:
        pool_pbar_m = list(
            tqdm(
                p.imap(kat.get_smoothed_vector,
                       pbar_m_list),
                total=len(pbar_m_list)
            )
        )    
```

```python
if pre_load == False:
    with Pool(max_pool) as p:
        pool_pbar_c = list(
            tqdm(
                p.imap(kat.get_smoothed_vector,
                       pbar_c_list),
                total=len(pbar_c_list)
            )
        )    
```

### Step 8
for each of your smoothed particle densities calculate the KS statistic from comparing the mutant to our altered control and from comparing the control to our altered control. Then calculate the difference between these two KS statistics by subtracting one from the other. Save the two KS statistics and the difference between the ks statistics within 3 lists.

```python
# Find the correlation between all of the controls and the samples.
# Note that many of the lowest values spring from very short transcripts or transcripts with few reads. 
if pre_load == False:
    ks_m_list = []
    ks_c_list = []
    ks_diff_list = []
    for pbar_m, pbar_c, alt_p in zip(pool_pbar_m, pool_pbar_c, pool_alt_p):
        sm_m, cumul_m = pbar_m
        sm_c, cumul_c = pbar_c
        sm_p, cumul_p = alt_p
        for d, i in zip(sm_m, list(range(len(sm_m)))):
            if d < 0:
                sm_m[i] = 0.000000000000000000001
        ks_m = max(abs(cumul_p - cumul_m))
        ks_c = max(abs(cumul_p - cumul_c))
        ks_diff = ks_c - ks_m
        ks_m_list.append(ks_m)
        ks_c_list.append(ks_c)
        ks_diff_list.append(ks_diff)
            
    # reorganizing the lists into a dataframe. 
    df_t = pd.DataFrame(list(zip(gene_list,ks_m_list, ks_c_list, ks_diff_list)))
    df_t.columns = ["gene_ID","ks_mutant", "ks_control", "ks_diff"]
    df_t.to_csv(save_path + "TASEP_ks_table_" + samp_num + ".csv", index=False)
```
